```python
import pandas as pd
import numpy as np

df = pd.read_csv('fatal-police-shootings-data.csv')
df.head()
```

```python
summary_table = pd.crosstab(df['race'], df['signs_of_mental_illness'])
summary_table
```

```python
summary_table['Percentage of mental illness (%)'] = summary_table.apply(lambda row: round((row[True] / (row[True] + row[False])) * 100, 2), axis=1)
summary_table
```

```python
# Rasa z największym odsetkiem znamion choroby psychicznej podczas interwencji
print("Rasa z największym odsetkiem znamion choroby psychicznej podczas interwencji:")
summary_table['Percentage of mental illness (%)'].idxmax()
```

```python
import matplotlib.pyplot as plt

df['date'] = pd.to_datetime(df['date'])

df['Day of the week'] = df['date'].dt.day_name()
df.head(10)
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['Day of the week'] = pd.Categorical(df['Day of the week'], categories=days, ordered=True)

interventions_day = df['Day of the week'].value_counts().sort_index()
interventions_day
```

```python
# Wykres
plt.figure(figsize=(10, 5))
interventions_day.plot(kind='bar', color='blue', edgecolor='black')

plt.xlabel('Day of the week')
plt.ylabel('Interventions count')
plt.title('Interventions count by day of the week')
plt.xticks(rotation=25)

plt.show()
```
