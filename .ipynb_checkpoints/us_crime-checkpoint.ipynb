```python
import pandas as pd
import numpy as np

df = pd.read_csv('fatal-police-shootings-data.csv')
df.head()
```

```python
summary_table = pd.crosstab(df['race'], df['signs_of_mental_illness'])
summary_table
```

```python
summary_table['Percentage of mental illness (%)'] = summary_table.apply(lambda row: round((row[True] / (row[True] + row[False])) * 100, 2), axis=1)
summary_table
```

```python
# Rasa z największym odsetkiem znamion choroby psychicznej podczas interwencji
print("Rasa z największym odsetkiem znamion choroby psychicznej podczas interwencji:")
summary_table['Percentage of mental illness (%)'].idxmax()
```

```python
import matplotlib.pyplot as plt

df['date'] = pd.to_datetime(df['date'])

df['Day of the week'] = df['date'].dt.day_name()
df.head(10)
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['Day of the week'] = pd.Categorical(df['Day of the week'], categories=days, ordered=True)

interventions_day = df['Day of the week'].value_counts().sort_index()
interventions_day
```

```python
# Wykres
plt.figure(figsize=(10, 5))
interventions_day.plot(kind='bar', color='blue', edgecolor='black')

plt.xlabel('Day of the week')
plt.ylabel('Interventions count')
plt.title('Interventions count by day of the week')
plt.xticks(rotation=25)

plt.show()
```

```python
popdata = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population', header=0)
statedata = pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._state_and_territory_abbreviations', header=0)

dfpop = popdata[0]
dfstate = statedata[1]

dfpop.head(15)
```

```python
dfstate.head(15)
```

```python
dfstate = dfstate[['Name', 'ANSI']].rename(columns={'Name': 'State', 'ANSI': 'State_Abbr'})
dfstate.head(15)
```

```python
df_merged = dfpop.merge(dfstate, on='State', how='left')
df_merged.head(15)
```

```python
df_final = df.merge(df_merged, left_on='state', right_on='State_Abbr', how='left')
df_final.head(15)
```

```python
incidents_per_state = df_final['state'].value_counts().reset_index()
incidents_per_state.head(15)
```

```python
incidents_per_state.columns = ['State_Abbr', 'Incidents']
incidents_per_state.head(15)
```

```python
df_crimes = df_merged.merge(incidents_per_state, on='State_Abbr', how='left').fillna(0)
df_crimes.head(15)
```

```python
df_crimes['Incidents_per_1000'] = ((df_crimes['Incidents'] / df_crimes['Census population, April 1, 2020 [1][2]']) * 1000)
df_crimes.head(15)
```

```python
df_crimes.sort_values('Incidents_per_1000', ascending=False).head(15)
```
